### 2)  Identify Nominal, Ordinal, Binary and Numeric attributes from data sets and display all values.



```python
nominal = ["CLIENTNUM", "Marital_Status", "Card_Category"]
ordinal = ["Education_Level", "Income_Category"]
binary = ["Attrition_Flag", "Gender"]
numeric = ["Customer_Age", "Dependent_count", "Months_on_book", 
"Credit_Limit","Total_Revolving_Bal", "Avg_Open_To_Buy", "Total_Amt_Chng_Q4_Q1","Total_Trans_Amt", "Total_Trans_Ct", "Total_Ct_Chng_Q4_Q1", "Avg_Utilization_Ratio"
]
print("Nominal",nominal)
print("Ordinal",ordinal)
print("Binary",binary)
print("Numeric",numeric)
```

### 3)  Identify symmetric and asymmetric binary attributes from data sets and display all values.



```python
print("Attrition_Flag values (Asymmetric Binary):")
print(df['Attrition_Flag'].value_counts())

print("Gender values (Symmetric Binary):")
print(df['Gender'].value_counts())
```

### 4)  For each quantitative attribute, calculate its average, standard deviation, minimum, mode, range and maximum values.



```python
quantitative_attributes = [
    "Customer_Age", "Dependent_count", "Months_on_book", "Credit_Limit",
    "Total_Revolving_Bal", "Avg_Open_To_Buy", "Total_Amt_Chng_Q4_Q1",
    "Total_Trans_Amt", "Total_Trans_Ct", "Total_Ct_Chng_Q4_Q1", "Avg_Utilization_Ratio"
]
for i in quantitative_attributes:
    print(':::',i,':::',sep="|")
    print("   Mean",df[i].mean())
    print("   Min",df[i].min())
    print("   Mode",df[i].mode())
    print("   Max",df[i].max())
    print("   Standard Deviation",df[i].std())
    print("   Ranging",df[i].max()-1)
```
